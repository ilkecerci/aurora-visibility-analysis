```python
df = df[df['time_diff'] < pd.Timedelta(days=30)]
```

```python
# Step1. Filtering geographically and statisticly
# Aurora Borealis can be seen only in Northern countries so I picked latitude as 55
# And since Iceland has a gap with more than 400 days, it's better to consider filtering a country if it has at least 600 days because Iceland has 649 unique scales. 
valid_countries = df['country'].value_counts()[df['country'].value_counts() >= 600].index

df_gold = df[
    (df['country'].isin(valid_countries)) & 
    (df['latitude'] >= 55)
].copy()

# I use .clip so the highest score doesn't go upper than 100
df_gold['visibility_score'] = (df_gold['visibility_km'] / 10).clip(upper=1) * 100
df_gold['darkness_score'] = 100 - df_gold['moon_illumination']

# Making the time data understandable for the machine
df_gold['last_updated'] = pd.to_datetime(df_gold['last_updated'])
df_gold['month'] = df_gold['last_updated'].dt.month

df_gold['sunset_dt'] = pd.to_datetime(df_gold['sunset'], format='%I:%M %p', errors='coerce')
df_gold['sunrise_dt'] = pd.to_datetime(df_gold['sunrise'], format='%I:%M %p', errors='coerce')

# Night Duration (hours)
df_gold['night_duration'] = (df_gold['sunrise_dt'] - df_gold['sunset_dt']).dt.total_seconds() / 3600
df_gold.loc[df_gold['night_duration'] < 0, 'night_duration'] += 24

# Identifying an advanced score (Non-linear)
# Cloudiness penalty (logarithmic)
df_gold['cloud_comp'] = np.exp(-df_gold['cloud'] / 25) * 100

# Night duration (Logarithmic): Increase of scoring would decrease after 12 hours of night time because more night-time is not equal to better observability (
df_gold['duration_comp'] = (np.log1p(df_gold['night_duration']) / np.log1p(14)) * 100

''' MODEL 1 (OLD): Linear Approach
This model aggregates each component with equal or fixed weight. 
Problem: Even if cloud cover is 100%, it will still be labeled as “visible” if the other factors are high (very incorrect).
Cloudiness is very important for Aurora Borealis!
df_gold['scientific_aurora_score'] = (
    (100 - df_gold['cloud']) * 0.40 + 
    ((df_gold['visibility_km'] / 10).clip(upper=1) * 100) * 0.20 + 
    (100 - df_gold['moon_illumination']) * 0.20 + 
    ((df_gold['night_duration'] / 12).clip(upper=1) * 100) * 0.20'''

# Final Advanced Score
df_gold['advanced_score'] = (
    (df_gold['cloud_comp'] * 0.60) +      # Cloudiness %60
    (df_gold['darkness_score'] * 0.15) +  # Moonlight %15
    (df_gold['visibility_score'] * 0.10) + # Visibility: %10
    (df_gold['duration_comp'] * 0.15)      # Night Duration: %15
).clip(upper=98) #Upper is not 100 because this dataset is not big enough to determine perfect.

# Final Table! Champs League
league_table = df_gold.groupby('country')['advanced_score'].agg(['mean', 'std', 'count', 'max'])
league_table = league_table.sort_values(by='mean', ascending=False)

print("Filters were implemented")
print(f"Number of chosen countries after filtering: {df_gold['country'].nunique()}")
print(f"Chosen Countries: {df_gold['country'].unique().tolist()}")
print("\nLeaderboard of Best Aurora Observing Countries")
print(league_table)
```

```python
# Let's compile the sensitivity analysis results
#Scenario 1: Cloud-Focused (Cloud: 0.7, Others: 0.1)
df_gold['score_cloud_heavy'] = (
    (df_gold['cloud_comp'] * 0.70) + 
    (df_gold['darkness_score'] * 0.10) + 
    (df_gold['visibility_score'] * 0.10) + 
    (df_gold['duration_comp'] * 0.10)
).clip(upper=98)

# Scenario 2: Darkness Focused (Darkness & Duration: 0.4, Cloud: 0.1, Visibility: 0.1)
df_gold['score_dark_heavy'] = (
    (df_gold['cloud_comp'] * 0.10) + 
    (df_gold['darkness_score'] * 0.40) + 
    (df_gold['visibility_score'] * 0.10) + 
    (df_gold['duration_comp'] * 0.40)
).clip(upper=98)

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Country-based means
sens_data = df_gold.groupby('country')[['advanced_score', 'score_cloud_heavy', 'score_dark_heavy']].mean()

# Graph drawing
plt.figure(figsize=(12, 7))
sens_data.sort_values('advanced_score', ascending=False).plot(kind='bar', figsize=(12, 7), width=0.8, color=['#6a0dad', '#2e8b57', '#ffa500'])

plt.title('Sensitivity Analysis: How Weight Shifts Affect Rankings', fontsize=14, fontweight='bold')
plt.ylabel('Average Score (0-98)', fontsize=12)
plt.xlabel('Country', fontsize=12)
plt.legend(['Original Model', 'Cloud Heavy (70%)', 'Darkness Heavy (80%)'], bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.savefig('../aurora_analysis/visuals_aurora/sensitivity_results.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Updated Geographical and Statistical Filtering
# I thought that maybe Norway and Finland has less data so I should lower the threshold to 365 (1 full year) instead of 600 days to include Norway and Finland.
valid_countries = df['country'].value_counts()[df['country'].value_counts() >= 365].index
df_gold = df[(df['country'].isin(valid_countries)) & (df['latitude'] >= 55)].copy()

# Continuity (30-day Gap Filter)
# Keeping the continuity rule to ensure the year is "complete" and not fragmented
df_gold['month'] = df_gold['last_updated'].dt.month
def check_continuity(group):
    group = group.sort_values('last_updated')
    gaps = group['last_updated'].diff().dt.days
    return gaps.max() <= 30

continuous_countries = df_gold.groupby('country').filter(check_continuity)['country'].unique()
df_gold = df_gold[df_gold['country'].isin(continuous_countries)].copy()

# Re-calculating the Scientific Score (Scientific Weights)
# Cloud (60%) [Exponential Penalty], Darkness (15%), Night (15%) [Logarithmic], Vis (10%)
df_gold['cloud_comp'] = np.exp(-df_gold['cloud'] / 25) * 100
df_gold['darkness_score'] = 100 - df_gold['moon_illumination']
df_gold['visibility_score'] = (df_gold['visibility_km'] / 10).clip(upper=1) * 100

# Night Duration calculation
df_gold['sunset_dt'] = pd.to_datetime(df_gold['sunset'], format='%I:%M %p', errors='coerce')
df_gold['sunrise_dt'] = pd.to_datetime(df_gold['sunrise'], format='%I:%M %p', errors='coerce')
df_gold['night_duration'] = (df_gold['sunrise_dt'] - df_gold['sunset_dt']).dt.total_seconds() / 3600
df_gold.loc[df_gold['night_duration'] < 0, 'night_duration'] += 24
df_gold['duration_comp'] = (np.log1p(df_gold['night_duration']) / np.log1p(14)) * 100

df_gold['advanced_score'] = (
    (df_gold['cloud_comp'] * 0.60) + (df_gold['darkness_score'] * 0.15) + 
    (df_gold['visibility_score'] * 0.10) + (df_gold['duration_comp'] * 0.15)
).clip(upper=98)

# Re-determining the Top 3 
# Let's see if Norway or Finland made a comeback!
new_league = df_gold.groupby('country')['advanced_score'].mean().sort_values(ascending=False)
print("Updated Aurora League (365-day threshold):")
print(new_league.head(5))

# Final Comparison Graph
top_3_final = new_league.index[:3].tolist()
df_plot = df_gold[df_gold['country'].isin(top_3_final)]

plt.figure(figsize=(14, 7))
sns.lineplot(data=df_plot, x='month', y='advanced_score', hue='country', marker='o', linewidth=3)
plt.axvline(x=3, color='orange', linestyle='--', alpha=0.5, label='NASA Equinox (March)')
plt.axvline(x=9, color='orange', linestyle='--', alpha=0.5, label='NASA Equinox (Sept)')
plt.title('Final 1-Year Trend: Most Reliable Aurora Spots', fontsize=16)
plt.xticks(range(1, 13))
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.savefig('../aurora_analysis/visuals_aurora/final_verification_365.png', dpi=300, bbox_inches='tight')
plt.show()
```
